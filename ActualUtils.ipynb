### Picture for Roles Selection

```python
import discord
from discord.ext import commands
from utils.var import *
from PIL import Image, ImageDraw, ImageFilter, ImageFont
import io
import requests


def welcome_image(text, text2):
    image = Image.open('./media/welcome.png', mode='r')

    width, height = image.size

    font = ImageFont.truetype('./media/Inter-Black-slnt=0.ttf', 300)
    #'./media/MonsieurLaDoulaise-Regular.ttf', 160)
    d = ImageDraw.Draw(image)
    msg = text2
    w1, h1 = d.textsize(msg, font=font)
    d.text(((width - w1) / 1, height - h1 - 10), msg, fill=(255, 255, 255), font=font)
    
    d = ImageDraw.Draw(image)
    msg = text
    w2, h2 = d.textsize(msg, font=font)
    d.text((width - w2, 0), msg, fill=(255, 255, 255), font=font)

    return image

text = ""
text2 = "<3"
welcome_image(text, text2).save("./data/tryingout.png")
welcome_image(text, text2)
```
